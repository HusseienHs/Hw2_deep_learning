```python
from collections import defaultdict
import pandas as pd
```

```python
import torch
import os

BASE_DIR = r"C:\Users\husseien\Desktop\340915149_322754953\Source Code"
DATA_DIR = os.path.join(BASE_DIR, "data")
TRAIN_CSV = os.path.join(DATA_DIR, "train.csv")
files_directory = os.path.join(DATA_DIR, "unlabeled", "unlabeled")

df =pd.read_csv(TRAIN_CSV)
df.sort_values(by=['id'], inplace=True)
to_do = list(zip(df['id'].tolist(), df['activity'].tolist()))
```

```python
# fill all ranges
ranges = defaultdict(list)

curr_activity = 'brushing_teeth'
curr_range_min = 0 
curr_range_max = 0

for i, (u_id,activity) in enumerate(to_do):
    if activity == curr_activity:
        curr_range_max = u_id
    else:
        ranges[curr_activity].append((curr_range_min, curr_range_max))
        curr_range_min = u_id
        curr_range_max = u_id
        curr_activity = activity
ranges[curr_activity].append((curr_range_min, curr_range_max))
```

```python
# fill all ranges
def find_activity(dictionary, id):
    for activity, ranges in dictionary.items():
        for start, end in ranges:
            if start <= id <= end:
                return activity
    return "AAAAAAAAAAAAAAAAAAAAAAAAAAAA"


BASE_DIR = r"C:\Users\husseien\Desktop\340915149_322754953\Source Code"
results_rf5 = r"C:\Users\husseien\Desktop\340915149_322754953\Source Code\results_rf5.csv"
res = pd.DataFrame(pd.read_csv(results_rf5)['sample_id'])

res['activity'] = res['sample_id'].map(lambda x: find_activity(ranges, x))
```
